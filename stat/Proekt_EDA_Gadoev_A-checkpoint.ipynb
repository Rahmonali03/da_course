```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('supermarket_sales - Sheet1.csv')
```

```python
df
```

```python
# Это среднее значение общей цены покупки (Total) для каждого филиала. 
# Такая метрика позволяет понять, какой филиал генерирует больше выручки по сравнению с другими.
average_total_by_branch = df.groupby('Branch')['Total'].mean()
print(average_total_by_branch)
```

```python
# Коэффициент корреляции между ценой (Unit price) и количеством (Quantity) товаров показывает, 
# существует ли линейная зависимость между этими двумя переменными.
correlation_price_quantity = df['Unit price'].corr(df['Quantity'])
print(correlation_price_quantity)
```

```python
# Сумма налоговых сборов (Tax 5%) для каждого типа клиентов (Customer type). Это позволяет увидеть, какой тип клиентов платит больше налогов.
total_tax_by_customer_type = df.groupby('Customer type')['Tax 5%'].sum()
print(total_tax_by_customer_type)
```

```python
# Меры размаха дают представление о распределении цен на товары, в том числе о минимальной, максимальной цене и медиане.
price_range = df['Unit price'].agg(['min', 'max', 'median'])
print(price_range)
```

```python
# Среднее количество купленных товаров (Quantity) для каждой категории товаров (Product line).
avg_quantity_by_product_line = df.groupby('Product line')['Quantity'].mean()
print(avg_quantity_by_product_line)
```

```python
# Этот коэффициент корреляции измеряет, существует ли связь между общими расходами (Total) и рейтингом покупателя (Rating). 
# Это может помочь понять, влияет ли сумма покупки на оценку покупателя.
correlation_total_rating = df['Total'].corr(df['Rating'])
print(correlation_total_rating)
```

```python
# Это средняя цена покупки (Total) для разных методов оплаты (Payment). 
# Это позволяет увидеть, какой метод оплаты используется для покупок большего размера.
avg_total_by_payment_method = df.groupby('Payment')['Total'].mean()
print(avg_total_by_payment_method)
```

```python
plt.figure(figsize=(15,17))
sns.boxplot(df)
# эти выбросы имеют аналитические ценности, поэтому их оставим нетронутыми
```

```python
# Гистограмма для распределения цен на товары
plt.figure(figsize=(8, 6))
sns.histplot(df['Unit price'], bins=30, kde=True, color='skyblue')

plt.title('Распределение цен товаров')
plt.xlabel('Цена товара ($)')
plt.ylabel('Частота')
plt.show()

```

```python
# Столбчатая диаграмма для сравнения средней общей стоимости по типу клиента
plt.figure(figsize=(8, 6))
sns.barplot(data=df, x='Customer type', y='Total', estimator='mean', ci=None)

plt.title('Средняя общая стоимость по типу клиента')
plt.xlabel('Тип клиента')
plt.ylabel('Средняя общая стоимость ($)')
plt.show()
```

```python
# Точечная диаграмма для изучения зависимости между количеством и общей ценой
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Quantity', y='Total', hue='Customer type', palette='viridis')

plt.title('Зависимость между количеством товаров и общей ценой')
plt.xlabel('Количество товаров')
plt.ylabel('Общая цена ($)')
plt.show()
```

```python
# Диаграмма размаха для оценки разброса валового дохода по типу клиента
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='Customer type', y='gross income')

plt.title('Разброс валового дохода по типу клиента')
plt.xlabel('Тип клиента')
plt.ylabel('Валовое поступление ($)')
plt.show()
```

```python
# Диаграмма размаха для оценки налога по типу оплаты
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='Payment', y='Tax 5%')

plt.title('Разброс налога в зависимости от типа оплаты')
plt.xlabel('Тип оплаты')
plt.ylabel('Налог 5% ($)')
plt.show()
```
