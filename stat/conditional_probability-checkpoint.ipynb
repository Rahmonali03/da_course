–≠—Ç–∏ —Å–æ–±—ã—Ç–∏—è –Ω–µ–∑–∞–≤–∏—Å–∏–º—ã, —Ç–∞–∫ –∫–∞–∫ —Ä–µ–∑—É–ª—å—Ç–∞—Ç –±—Ä–æ—Å–∫–∞ –º–æ–Ω–µ—Ç—ã –Ω–µ –≤–ª–∏—è–µ—Ç –Ω–∞ —Ä–µ–∑—É–ª—å—Ç–∞—Ç –±—Ä–æ—Å–∫–∞ –∫—É–±–∏–∫–∞.  

### –ó–∞–≤–∏—Å–∏–º—ã–µ —Å–æ–±—ã—Ç–∏—è

–û–ø—Ä–µ–¥–µ–ª–∏—Ç–µ –∫–æ—Ç–æ—Ä—ã–µ –∏–∑ –ø—Ä–∏–≤–µ–¥—ë–Ω–Ω—ã—Ö —Å–æ–±—ã—Ç–∏–π —è–≤–ª—è—é—Ç—Å—è –Ω–µ–∑–∞–≤–∏—Å–∏–º—ã–º–∏ –ª–∏–±–æ –∑–∞–≤–∏—Å–∏–º—ã–º–∏:

* –≤ —Ä—É–ª–µ—Ç–∫–µ –¥–≤–∞ —Ä–∞–∑–∞ –ø–æ–¥—Ä—è–¥ –≤—ã–ø–∞–ª–æ 20 —á—ë—Ä–Ω–æ–µ | <input type="checkbox"> *–Ω–µ–∑–∞–≤–∏—Å–∏–º—ã–µ* <input type="checkbox"> *–∑–∞–≤–∏—Å–∏–º—ã–µ*  
* –≤—ã –Ω–µ –ø–æ–¥–∞–ª–∏ –º–∏–ª–æ—Å—Ç—ã–Ω—é –∏ —É –≤–∞—Å —Å–ª–æ–º–∞–ª–∞—Å—å –º–∞—à–∏–Ω–∞ | <input type="checkbox"> *–Ω–µ–∑–∞–≤–∏—Å–∏–º—ã–µ* <input type="checkbox"> *–∑–∞–≤–∏—Å–∏–º—ã–µ*  
* –≤—ã —É–ª—É—á—à–∏–ª–∏ –¥–∏–∑–∞–π–Ω –ø—Ä–æ–¥—É–∫—Ü–∏–∏ –∏ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø—Ä–æ–¥–∞–∂ —É–≤–µ–ª–∏—á–∏–ª–æ—Å—å | <input type="checkbox"> *–Ω–µ–∑–∞–≤–∏—Å–∏–º—ã–µ* <input type="checkbox"> *–∑–∞–≤–∏—Å–∏–º—ã–µ*    

## –û–±—â–∞—è —Ñ–æ—Ä–º—É–ª–∞ —É–º–Ω–æ–∂–µ–Ω–∏—è

**–û–±—â–∞—è —Ñ–æ—Ä–º—É–ª–∞ —É–º–Ω–æ–∂–µ–Ω–∏—è** - –¥–ª—è –ª—é–±—ã—Ö –¥–≤—É—Ö —Å–æ–±—ã—Ç–∏–π $A$ –∏ $B$: $$P(A \cap B)=P(A) \cdot P(B|A)$$ –≥–¥–µ:  
* $P(A)$ - –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –Ω–∞—Å—Ç—É–ø–ª–µ–Ω–∏—è —Å–æ–±—ã—Ç–∏—è $A$,  
* $P(B|A)$ - –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –Ω–∞—Å—Ç—É–ø–ª–µ–Ω–∏—è —Å–æ–±—ã—Ç–∏—è $B$, –ø—Ä–∏ —É—Å–ª–æ–≤–∏–∏ —á—Ç–æ –Ω–∞—Å—Ç—É–ø–∏–ª–æ —Å–æ–±—ã—Ç–∏–µ $A$.  



---

>–û–±—â–µ–µ –ø—Ä–∞–≤–∏–ª–æ —Å–ª–æ–∂–µ–Ω–∏—è: $P(A \cup B)=P(A)+P(B)-P(A \cap B)$  
>–û–±—â–µ–µ –ø—Ä–∞–≤–∏–ª–æ —É–º–Ω–æ–∂–µ–Ω–∏—è: $P(A \cap B)=P(A) \cdot P(B|A)$

---



**–ü—Ä–∏–º–µ—Ä**  
–í —Ç–∞–±–ª–∏—Ü–µ –Ω–∏–∂–µ –ø—Ä–∏–≤–æ–¥–∏—Ç—Å—è —Ä–µ–∑—É–ª—å—Ç–∞—Ç –æ–ø—Ä–æ—Å–∞ —Å—Ç—É–¥–µ–Ω—Ç–æ–≤ –Ω–∞ –ø—Ä–µ–¥–º–µ—Ç —Ç–æ–≥–æ —É–º–µ—é—Ç –ª–∏ –æ–Ω–∏ –≤–æ–¥–∏—Ç—å –º–∞—à–∏–Ω—É.

|             | –£–º–µ–µ—Ç –≤–æ–¥–∏—Ç—å (D) | –ù–µ —É–º–µ–µ—Ç –≤–æ–¥–∏—Ç—å (D') | –í—Å–µ–≥–æ   |
|-------------|------------------|----------------------|---------|
| –î–µ–≤—É—à–∫–∞ (F) | 36               | 144                  | **180** |  
| –ü–∞—Ä–µ–Ω—å  (M) | 288              | 32                   | **320** |  
| **–í—Å–µ–≥–æ**   | **324**          | **176**              | **500** |

–î–æ–ø—É—Å—Ç–∏–º, —á—Ç–æ –º—ã —Ä–∞–Ω–¥–æ–º–Ω–æ –≤—ã–±–∏—Ä–∞–µ–º –∏–∑ —ç—Ç–æ–π –≥—Ä—É–ø–ø—ã –æ–¥–Ω–æ–≥–æ —Å—Ç—É–¥–µ–Ω—Ç–∞. –ü–æ—Å—Ç–∞—Ä–∞–µ–º—Å—è –Ω–∞–π—Ç–∏ –æ—Ç–≤–µ—Ç—ã –Ω–∞ —Å–ª–µ–¥—É—é—â–∏–µ –≤–æ–ø—Ä–æ—Å—ã:  

a) –∫–∞–∫–æ–≤–∞ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å, —á—Ç–æ —ç—Ç–æ—Ç —Å—Ç—É–¥–µ–Ω—Ç —É–º–µ–µ—Ç –≤–æ–¥–∏—Ç—å? –í—Å–µ–≥–æ —É –Ω–∞—Å 500 —Å—Ç—É–¥–µ–Ω—Ç–æ–≤, 324 –∏–∑ –∫–æ—Ç–æ—Ä—ã—Ö —É–º–µ—é—Ç –≤–æ–¥–∏—Ç—å. –°–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ, $P(D)=\frac{324}{500}=.648$  

b) –∫–∞–∫–æ–≤–∞ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å, —á—Ç–æ —ç—Ç–æ—Ç —Å—Ç—É–¥–µ–Ω—Ç –ø–∞—Ä–µ–Ω—å? –í—Å–µ–≥–æ –≤ –≥—Ä—É–ø–ø–µ 320 –ø–∞—Ä–Ω–µ–π, –∞ –∑–Ω–∞—á–∏—Ç $P(M)=\frac{320}{500}=.64$  

c) –∫–∞–∫–æ–≤–∞ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å, —á—Ç–æ —ç—Ç–æ—Ç —Å—Ç—É–¥–µ–Ω—Ç –ø–∞—Ä–µ–Ω—å –∏ –æ–Ω —É–º–µ–µ—Ç –≤–æ–¥–∏—Ç—å? 288 –ø–∞—Ä–Ω–µ–π —É–º–µ—é—Ç –≤–æ–¥–∏—Ç—å, —Ç–∞–∫–∏–º –æ–±—Ä–∞–∑–æ–º $P(M \cap D)=\frac{288}{500}=.576$  

d) –∫–∞–∫–æ–≤–∞ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ —ç—Ç–æ—Ç —Å—Ç—É–¥–µ–Ω—Ç —É–º–µ–µ—Ç –≤–æ–¥–∏—Ç—å, –ø—Ä–∏ —É—Å–ª–æ–≤–∏–∏ —á—Ç–æ –æ–Ω –ø–∞—Ä–µ–Ω—å? –í—Å–µ–≥–æ –ø–∞—Ä–Ω–µ–π 320, –∏–∑ –∫–æ—Ç–æ—Ä—ã—Ö 288 —É–º–µ–µ—Ç –≤–æ–¥–∏—Ç—å. –°–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ, $P(D|M)=\frac{288}{320}=.9$

–ê —Ç–µ–ø–µ—Ä—å –ø—Ä–µ–¥—Å—Ç–∞–≤–∏–º, —á—Ç–æ –Ω–∞–º –¥–∞–Ω—ã –ª–∏—à—å –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Å–æ–±—ã—Ç–∏—è $P(M)$ –∏ $P(D|M)$ –∏ –Ω–∞–º –Ω—É–∂–Ω–æ —Ä–∞—Å—Å—á–∏—Ç–∞—Ç—å –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å $P(M \cap D)$. –ò—Å–ø–æ–ª—å–∑—É—è –æ–±—â—É—é —Ñ–æ—Ä–º—É–ª—É —É–º–Ω–æ–∂–µ–Ω–∏—è –ø–æ–ª—É—á–∞–µ–º:  

$$P(M \cap D) = P(M) \cdot P(D|M)=.64 \cdot .9 = .576$$  
–ú—ã –ø–æ–ª—É—á–∏–ª–∏ —Ç–æ—Ç –∂–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç, —á—Ç–æ –∏ –ø—Ä–∏ –æ—Ç–≤–µ—Ç–µ –Ω–∞ –≤–æ–ø—Ä–æ—Å **c)**, –Ω–æ –∏–º–µ—è —Ç–æ–ª—å–∫–æ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ –¥–≤—É—Ö –¥—Ä—É–≥–∏—Ö —Å–æ–±—ã—Ç–∏–π.

## –£—Å–ª–æ–≤–Ω–∞—è –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å

–ü–µ—Ä–µ—Å—Ç–∞–≤–∏–≤ –º–µ—Å—Ç–∞–º–∏ —ç–ª–µ–º–µ–Ω—Ç—ã –æ–±—â–µ–π —Ñ–æ—Ä–º—É–ª—ã —É–º–Ω–æ–∂–µ–Ω–∏—è –º—ã –ø–æ–ª—É—á–∏–º —Ñ–æ—Ä–º—É–ª—É —É—Å–ª–æ–≤–Ω–æ–π –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏:  
  
$$P(B|A)=\frac{P(A \cap B)}{P(A)}$$

üí°–û–±—Ä–∞—Ç–∏—Ç–µ –≤–Ω–∏–º–∞–Ω–∏–µ, —á—Ç–æ –≤ –¥–µ–Ω–æ–º–∏–Ω–∞—Ç–æ—Ä–µ –≤—Å–µ–≥–¥–∞ —Å—Ç–æ–∏—Ç –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Å–æ–±—ã—Ç–∏—è "–ø—Ä–∏ —É—Å–ª–æ–≤–∏–∏".

–£—Å–ª–æ–≤–Ω–∞—è –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –∏—Å–ø–æ–ª—å–∑—É–µ—Ç—Å—è –∫–æ–≥–¥–∞ –¥–≤–∞ –∏ –±–æ–ª–µ–µ —Å–æ–±—ã—Ç–∏–π –Ω–µ —è–≤–ª—è—é—Ç—Å—è –Ω–µ–∑–∞–≤–∏—Å–∏–º—ã–º–∏. –≠—Ç–æ –æ–∑–Ω–∞—á–∞–µ—Ç —á—Ç–æ –Ω–∞ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –æ–¥–Ω–æ–≥–æ —Å–æ–±—ã—Ç–∏—è –≤–ª–∏—è–µ—Ç —Ç–æ, –ø—Ä–æ–∏–∑–æ—à–ª–æ –ª–∏ –¥—Ä—É–≥–æ–µ —Å–æ–±—ã—Ç–∏–µ.



---
‚ùì–ó–∞–¥–∞—á–∞  
–ö–ª–∏–Ω–∏—á–µ—Å–∫–∏–µ –∏—Å—Å–ª–µ–¥–æ–≤–∞–Ω–∏—è –æ–¥–Ω–æ–≥–æ –∞–Ω—Ç–∏–¥–µ–ø—Ä–µ—Å—Å–∞–Ω—Ç–∞ –ø–æ–∫–∞–∑–∞–ª–∏, —á—Ç–æ –ø—Ä–∏ —É–ø–æ—Ç—Ä–µ–±–ª–µ–Ω–∏–∏ –¥–∞–Ω–Ω–æ–≥–æ –ø—Ä–µ–ø–∞—Ä–∞—Ç–∞ –∏–º–µ–µ—Ç—Å—è 14% –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ –ø–æ–±–æ—á–Ω–æ–≥–æ —ç—Ñ—Ñ–µ–∫—Ç–∞ –≤ –≤–∏–¥–µ –±–µ—Å—Å–æ–Ω–Ω–∏—Ü—ã (I), 26% –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ –ø–æ–±–æ—á–Ω–æ–≥–æ —ç—Ñ—Ñ–µ–∫—Ç–∞ –≤ –≤–∏–¥–µ –≥–æ–ª–æ–≤–Ω—ã—Ö –±–æ–ª–µ–π (H), –∏ 5% –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ –æ–±–æ–∏—Ö –ø–æ–±–æ—á–Ω—ã—Ö —ç—Ñ—Ñ–µ–∫—Ç–æ–≤ (I and H).  

* –ü—Ä–µ–¥–ø–æ–ª–æ–∂–∏–º, —á—Ç–æ –ø–∞—Ü–∏–µ–Ω—Ç –ø—Ä–∏–Ω–∏–º–∞–≤—à–∏–π —ç—Ç–æ—Ç –ø—Ä–µ–ø–∞—Ä–∞—Ç –∏—Å–ø—ã—Ç—ã–≤–∞–µ—Ç –±–µ—Å—Å–æ–Ω–Ω–∏—Ü—É. –ö–∞–∫–æ–≤–∞ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ –æ–Ω —Ç–∞–∫–∂–µ –±—É–¥–µ—Ç –∏—Å–ø—ã—Ç—ã–≤–∞—Ç—å –≥–æ–ª–æ–≤–Ω—ã–µ –±–æ–ª–∏?  

* –ü—Ä–µ–¥–ø–æ–ª–æ–∂–∏–º, —á—Ç–æ –ø—Ä–µ–ø–∞—Ä–∞—Ç –≤—ã–∑—ã–≤–∞–µ—Ç —É –ø–∞—Ü–∏–µ–Ω—Ç–∞ –≥–æ–ª–æ–≤–Ω—ã–µ –±–æ–ª–∏. –ö–∞–∫–æ–≤–∞ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ –æ–Ω —Ç–∞–∫–∂–µ –±—É–¥–µ—Ç –≤—ã–∑—ã–≤–∞—Ç—å –±–µ—Å—Å–æ–Ω–Ω–∏—Ü—É?


```python
p_insomnia = .14
p_headache = .26
p_insomnia_and_headache = .05

p_headache_given_insomnia = p_insomnia_and_headache / p_insomnia
p_insomnia_given_headache = p_insomnia_and_headache / p_headache

print(f"–≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –≥–æ–ª–æ–≤–Ω–æ–π –±–æ–ª–∏ –ø—Ä–∏ —É—Å–ª–æ–≤–∏–∏ –±–µ—Å—Å–æ–Ω–Ω–∏—Ü—ã: {round(p_headache_given_insomnia,2)}")
print(f"–≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –±–µ—Å—Å–æ–Ω–Ω–∏—Ü—ã –ø—Ä–∏ —É—Å–ª–æ–≤–∏–∏ –≥–æ–ª–æ–≤–Ω–æ–π –±–æ–ª–∏: {round(p_insomnia_given_headache, 2)}")
```

## –ù–∞—Ö–æ–∂–¥–µ–Ω–∏–µ —É—Å–ª–æ–≤–Ω–æ–π –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ —Ç–∞–±–ª–∏—á–Ω—ã–º —Å–ø–æ—Å–æ–±–æ–º

–¢–µ–ø–µ—Ä—å –ø–æ–ø—Ä–æ–±—É–µ–º –Ω–∞–π—Ç–∏ –æ—Ç–≤–µ—Ç –Ω–∞ —Å–ª–µ–¥—É—é—â–∏–π –≤–æ–ø—Ä–æ—Å: –µ—Å–ª–∏ –ø—Ä–µ–ø–∞—Ä–∞—Ç –≤—ã–∑—ã–≤–∞–µ—Ç —É –ø–∞—Ü–∏–µ–Ω—Ç–∞ –±–µ—Å—Å–æ–Ω–Ω–∏—Ü—É, –∫–∞–∫–æ–≤–∞ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —á—Ç–æ –æ–Ω –Ω–µ –±—É–¥–µ—Ç –∏—Å–ø—ã—Ç—ã–≤–∞—Ç—å –≥–æ–ª–æ–≤–Ω—ã–µ –±–æ–ª–∏?  

–ù–∞—Ö–æ–∂–¥–µ–Ω–∏–µ –æ—Ç–≤–µ—Ç–∞ –Ω–∞ —Ç–∞–∫–æ–π –≤–æ–ø—Ä–æ—Å –∑–∞–¥–∞—á–∞ –Ω–µ—Ç—Ä–∏–≤–∏–∞–ª—å–Ω–∞—è. –û–¥–∏–Ω –∏–∑ —Å–ø–æ—Å–æ–±–æ–≤, –∫–æ—Ç–æ—Ä—ã–π –º—ã –º–æ–∂–µ–º –¥–ª—è —ç—Ç–æ–≥–æ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å - —Ç–∞–±–ª–∏—á–Ω—ã–π —Å–ø–æ—Å–æ–±.  

–ü—Ä–µ–∂–¥–µ –≤—Å–µ–≥–æ, –≤—Å–ø–æ–º–Ω–∏–º –∫–∞–∫–∏–µ —É –Ω–∞—Å –µ—Å—Ç—å –∏—Å—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ:  
* $P(I)=.14$  
* $P(H)=.26$  
* $P(I \cap H)=.05$  

–ò—Å–ø–æ–ª—å–∑—É—è —ç—Ç–∏ –∏—Å—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ, –ø–æ—Å—Ç—Ä–æ–∏–º —Å–ª–µ–¥—É—é—â—É—é —Ç–∞–±–ª–∏—Ü—É:  

|           | I         | not_I | **total** |  
|-----------|-----------|-------|-----------|  
| H         | .05       |       | **.26**   |  
| not_H     |           |       |           |       
| **total** | **.14**   |       | **1**     |

–¢–∞–∫ –∫–∞–∫ –≤ –ø–æ—Å–ª–µ–¥–Ω–µ–º —Å—Ç–æ–ª–±—Ü–µ –∏ –≤ –ø–æ—Å–ª–µ–¥–Ω–µ–π —Å—Ç—Ä–æ–∫–µ —É –Ω–∞—Å –Ω–∞—Ö–æ–¥—è—Ç—Å—è –∏—Ç–æ–≥–∏, –ø—Ä–æ—Å—Ç—ã–º–∏ –æ–ø–µ—Ä–∞—Ü–∏—è–º–∏ –≤—ã—á–∏—Ç–∞–Ω–∏—è –º—ã –º–æ–∂–µ–º –∑–∞–ø–æ–ª–Ω–∏—Ç—å –ø—É—Å—Ç—ã–µ —è—á–µ–π–∫–∏. –í —Ä–µ–∑—É–ª—å—Ç–∞—Ç–µ –º—ã –ø–æ–ª—É—á–∏–º –∫–∞–∫ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ –∫–∞–∂–¥–æ–≥–æ –æ—Ç–¥–µ–ª—å–Ω–æ–≥–æ —Å–æ–±—ã—Ç–∏—è —Ç–∞–∫ –∏ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ —Ç–æ–≥–æ —á—Ç–æ –æ–Ω–∏ –ø—Ä–æ–∏–∑–æ–π–¥—É—Ç –≤–º–µ—Å—Ç–µ.  

|           | I         | not_I   | **total** |  
|-----------|-----------|---------|-----------|  
| H         | .05       | .21     | **.26**   |  
| not_H     | .09       | .65     | **.74**   |       
| **total** | **.14**   | **.86** | **1**     |

–ò—Å–ø–æ–ª—å–∑—É—è –∑–∞–ø–æ–ª–Ω–µ–Ω–Ω—É—é —Ç–∞–±–ª–∏—Ü—É –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–µ–π, –æ—Ç–≤–µ—Ç–∏–º –Ω–∞ —Å–ª–µ–¥—É—é—â–∏–µ –≤–æ–ø—Ä–æ—Å—ã:  

a) –µ—Å–ª–∏ –ø—Ä–µ–ø–∞—Ä–∞—Ç –≤—ã–∑—ã–≤–∞–µ—Ç —É –ø–∞—Ü–∏–µ–Ω—Ç–∞ –±–µ—Å—Å–æ–Ω–Ω–∏—Ü—É, –∫–∞–∫–æ–≤–∞ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —á—Ç–æ –æ–Ω –Ω–µ –±—É–¥–µ—Ç –∏—Å–ø—ã—Ç—ã–≤–∞—Ç—å –≥–æ–ª–æ–≤–Ω—ã–µ –±–æ–ª–∏?  
$$P(H'|I)=\frac{P(I \cap H')}{P(I)}=\frac{.09}{.14}‚âà.64$$  

b) –µ—Å–ª–∏ –ø—Ä–µ–ø–∞—Ä–∞—Ç –≤—ã–∑—ã–≤–∞–µ—Ç —É –ø–∞—Ü–∏–µ–Ω—Ç–∞ –≥–æ–ª–æ–≤–Ω—ã–µ –±–æ–ª–∏, –∫–∞–∫–æ–≤–∞ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —á—Ç–æ –æ–Ω –Ω–µ –±—É–¥–µ—Ç –∏—Å–ø—ã—Ç—ã–≤–∞—Ç—å –±–µ—Å—Å–æ–Ω–Ω–∏—Ü—É?  
$$P(I'|H)=\frac{P(H \cap I')}{P(H)}=\frac{.21}{.26}‚âà.81$$



---
‚ùì–ó–∞–¥–∞—á–∞  
–û–ø—Ä–µ–¥–µ–ª–∏—Ç–µ —Å–ª–µ–¥—É—é—â–∏–µ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏:  
a) –µ—Å–ª–∏ –ø—Ä–µ–ø–∞—Ä–∞—Ç –Ω–µ –≤—ã–∑—ã–≤–∞–µ—Ç —É –ø–∞—Ü–∏–µ–Ω—Ç–∞ –±–µ—Å—Å–æ–Ω–Ω–∏—Ü—É, –∫–∞–∫–æ–≤–∞ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —á—Ç–æ –æ–Ω –±—É–¥–µ—Ç –∏—Å–ø—ã—Ç—ã–≤–∞—Ç—å –≥–æ–ª–æ–≤–Ω—ã–µ –±–æ–ª–∏?  
b) –µ—Å–ª–∏ –ø—Ä–µ–ø–∞—Ä–∞—Ç –Ω–µ –≤—ã–∑—ã–≤–∞–µ—Ç —É –ø–∞—Ü–∏–µ–Ω—Ç–∞ –≥–æ–ª–æ–≤–Ω—ã–µ –±–æ–ª–∏, –∫–∞–∫–æ–≤–∞ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —á—Ç–æ –æ–Ω —Ç–∞–∫–∂–µ –Ω–µ –±—É–¥–µ—Ç –≤—ã–∑—ã–≤–∞—Ç—å –±–µ—Å—Å–æ–Ω–Ω–∏—Ü—É?

```python
p_insomnia = .14
p_headache = .26
p_insomnia_and_headache = .05

p_headache_and_not_insomnia = p_headache - p_insomnia_and_headache
# print(p_headache_and_not_insomnia)

p_not_insomnia = 1 - p_insomnia
p_headache_given_not_insomnia = p_headache_and_not_insomnia / p_not_insomnia

p_not_insomnia_and_not_headache = p_not_insomnia - p_headache_and_not_insomnia
p_not_insomnia_given_not_headache = p_not_insomnia_and_not_headache / p_not_headache

print(p_headache_given_not_insomnia)
print(p_not_insomnia_given_not_headache)
```

‚ùì–ó–∞–¥–∞—á–∞  
–ù–∞–º –∏–∑–≤–µ—Å—Ç–Ω—ã —Å–ª–µ–¥—É—é—â–∏–µ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏:  
* $P(B)=.4$  
* $P(D)=.95$  
* $P(B \cap D)=.38$  

–ù–∞ –æ—Å–Ω–æ–≤–µ —É–∫–∞–∑–∞–Ω–Ω—ã—Ö –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–µ–π –æ–ø—Ä–µ–¥–µ–ª–∏—Ç–µ —è–≤–ª—è—é—Ç—Å—è –ª–∏ —Å–æ–±—ã—Ç–∏—è B –∏ D –∑–∞–≤–∏—Å–∏–º—ã–º–∏.



$P(A \cap B)=P(A) \cdot P(B)$  
$P(A \cap B)=P(A) \cdot P(B|A)$

```python
p_b = .4
p_d = .95
p_b_and_d = .38

p_b_given_d = p_b_and_d / p_d
p_b_given_d == p_b
```

## –î–µ—Ä–µ–≤–æ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–µ–π

–í—Ç–æ—Ä–æ–π –º–µ—Ç–æ–¥ –Ω–∞—Ö–æ–∂–¥–µ–Ω–∏—è –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ —Å–æ–±—ã—Ç–∏–π —ç—Ç–æ `–¥–µ—Ä–µ–≤–æ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–µ–π`.

**–ü—Ä–∏–º–µ—Ä**  
–¢–æ—Ä–≥–æ–≤—ã–π –ø—Ä–µ–¥—Å—Ç–∞–≤–∏—Ç–µ–ª—å —Å–æ–æ–±—â–∞–µ—Ç —Å–≤–æ–µ–º—É –¥—Ä—É–≥—É, —á—Ç–æ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ —á—Ç–æ –æ–Ω –∑–∞–∫–ª—é—á–∏—Ç –≤–∞–∂–Ω—ã–π –∫–æ–Ω—Ç—Ä–∞–∫—Ç —Å–æ—Å—Ç–∞–≤–ª—è–µ—Ç 40%. –ï—Å–ª–∏ –∫–æ–Ω—Ç—Ä–∞–∫—Ç –±—É–¥–µ—Ç –∑–∞–∫–ª—é—á—ë–Ω, –æ–Ω –ø–æ–ª—É—á–∏—Ç –±–æ–ª—å—à–∏–µ –ø—Ä–µ–º–∏–∞–ª—å–Ω—ã–µ –∏ —Å –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å—é 90% –ø–æ–µ–¥–µ—Ç –≤ –æ—Ç–ø—É—Å–∫ –≤ –î—É–±–∞–π. –î–∞–∂–µ –µ—Å–ª–∏ –∫–æ–Ω—Ç—Ä–∞–∫—Ç –Ω–µ —É–¥–∞—Å—Ç—Å—è –∑–∞–∫–ª—é—á–∏—Ç—å –æ–Ω –≤—Å—ë —Ä–∞–≤–Ω–æ –ø–ª–∞–Ω–∏—Ä—É–µ—Ç –ø—Ä–æ–≤–µ—Å—Ç–∏ –æ—Ç–ø—É—Å–∫ –≤ –î—É–±–∞–µ, –Ω–æ —É–∂–µ —Å –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å—é 30%.  

–ò—Å–ø–æ–ª—å–∑—É—è –¥–∞–Ω–Ω—É—é –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é –ø–æ–ø—Ä–æ–±—É–µ–º –ø–æ—Å—Ç—Ä–æ–∏—Ç—å –¥–µ—Ä–µ–≤–æ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–µ–π.  

–ü–µ—Ä–≤—ã–π —É—Ä–æ–≤–µ–Ω—å –¥–µ—Ä–µ–≤–∞ —Å–æ—Å—Ç–æ–∏—Ç –∏–∑ –¥–≤—É—Ö –≤–µ—Ç–æ–∫. –í –∫–∞–∂–¥–æ–º –ø–æ—Å–ª–µ–¥—É—é—â–µ–º —É—Ä–æ–≤–Ω–µ –¥–µ—Ä–µ–≤–∞ –∏–∑ –≤–µ—Ç–æ–∫ –ø—Ä–µ–¥—ã–¥—É—â–∏—Ö —É—Ä–æ–≤–Ω–µ–π –≤—ã–≤–æ–¥–∏—Ç—Å—è –ø–∞—Ä–∞ –Ω–æ–≤—ã—Ö –≤–µ—Ç–æ–∫.  

–ü–µ—Ä–≤—ã–π —É—Ä–æ–≤–µ–Ω—å –Ω–∞—à–µ–≥–æ –¥–µ—Ä–µ–≤–∞ –±—É–¥–µ—Ç —Å–≤—è–∑–∞–Ω —Å –∫–æ–Ω—Ç—Ä–∞–∫—Ç–æ–º. –ù–∞–º –Ω—É–∂–Ω–æ –ø–æ—Å—Ç—Ä–æ–∏—Ç—å –¥–≤–µ –≤–µ—Ç–∫–∏ - –≤–µ—Ç–∫–∞ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ –∑–∞–∫–ª—é—á–µ–Ω–∏—è –∫–æ–Ω—Ç—Ä–∞–∫—Ç–∞ ($P(C)$) –∏ –≤–µ—Ç–∫–∞ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ –Ω–µ –∑–∞–∫–ª—é—á–µ–Ω–∏—è –∫–æ–Ω—Ç—Ä–∞–∫—Ç–∞ ($P(C')$). –ú—ã –∑–Ω–∞–µ–º $P(C)$, —Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ –º–æ–∂–µ–º –≤—ã—á–∏—Å–ª–∏—Ç—å –µ–≥–æ –∫–æ–º–ø–ª–µ–º–µ–Ω—Ç:  
* $P(C)=.4$  
* $P(C')=1-P(C)=.6$

–¢–µ–ø–µ—Ä—å, –¥–ª—è –∫–∞–∂–¥–æ–π –≤–µ—Ç–∫–∏ –ø–µ—Ä–≤–æ–≥–æ —É—Ä–æ–≤–Ω—è –Ω–∞–º –Ω—É–∂–Ω–æ –ø–æ—Å—Ç—Ä–æ–∏—Ç—å –≤–µ—Ç–∫–∏ —Å–ª–µ–¥—É—é—â–∏—Ö –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–µ–π:  
* –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –æ—Ç–ø—É—Å–∫–∞ –ø—Ä–∏ —É—Å–ª–æ–≤–∏–∏ –∑–∞–∫–ª—é—á–µ–Ω–∏—è –∫–æ–Ω—Ç—Ä–∞–∫—Ç–∞ $P(V|C)=.9$  
* –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –æ—Ç–º–µ–Ω—ã –æ—Ç–ø—É—Å–∫–∞ –ø—Ä–∏ —É—Å–ª–æ–≤–∏–∏ –∑–∞–∫–ª—é—á–µ–Ω–∏—è –∫–æ–Ω—Ç—Ä–∞–∫—Ç–∞ $P(V'|C)=1-P(V|C)=1-.9=.1$  

* –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –æ—Ç–ø—É—Å–∫–∞ –ø—Ä–∏ —É—Å–ª–æ–≤–∏–∏ —á—Ç–æ –∫–æ–Ω—Ç—Ä–∞–∫—Ç –Ω–µ –±—É–¥–µ—Ç –∑–∞–∫–ª—é—á–µ–Ω $P(V|C')=.3$

## –§–æ—Ä–º—É–ª–∞ –ø–æ–ª–Ω–æ–π –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏

–ò–º–µ—è –≥–æ—Ç–æ–≤–æ–µ –¥–µ—Ä–µ–≤–æ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–µ–π, –ø–æ–ø—Ä–æ–±—É–µ–º –Ω–∞–π—Ç–∏ –æ—Ç–≤–µ—Ç –Ω–∞ —Å–ª–µ–¥—É—é—â–∏–π –≤–æ–ø—Ä–æ—Å: *–∫–∞–∫–æ–≤–∞ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ —Ç–æ—Ä–≥–æ–≤—ã–π –ø—Ä–µ–¥—Å—Ç–∞–≤–∏—Ç–µ–ª—å –ø–æ–µ–¥–µ—Ç –≤ –æ—Ç–ø—É—Å–∫?*

–£ –Ω–∞—Å –µ—Å—Ç—å –¥–≤–∞ —Å—Ü–µ–Ω–∞—Ä–∏—è:  
* —Ç–æ—Ä–≥–æ–≤—ã–π –ø—Ä–µ–¥—Å—Ç–∞–≤–∏—Ç–µ–ª—å –∑–∞–∫–ª—é—á–∞–µ—Ç –∫–æ–Ω—Ç—Ä–∞–∫—Ç **–∏** –µ–¥–µ—Ç –≤ –æ—Ç–ø—É—Å–∫  
* **–∏–ª–∏** —Ç–æ—Ä–≥–æ–≤—ã–π –ø—Ä–µ–¥—Å—Ç–∞–≤–∏—Ç–µ–ª—å –Ω–µ –∑–∞–∫–ª—é—á–∞–µ—Ç –∫–æ–Ω—Ç—Ä–∞–∫—Ç **–∏** –µ–¥–µ—Ç –≤ –æ—Ç–ø—É—Å–∫

–°–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ: $P(V)=P(C \cap V) + P(C' \cap V)$

–ò—Å–ø–æ–ª—å–∑—É—è –¥–µ—Ä–µ–≤–æ —Å—á–∏—Ç–∞–µ–º:  
* $P(C \cap V)=P(C) \cdot P(V|C)=.4 \cdot .9=.36$  
* $P(C' \cap V)=P(C') \cdot P(V|C')=.6 \cdot .3=.18$  

–¢–∞–∫–∏–º –æ–±—Ä–∞–∑–æ–º:
