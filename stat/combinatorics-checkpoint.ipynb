# Комбинаторика

**Комбинаторика** — это раздел математики, который изучает способы формирования, выбора и расположения объектов в различных комбинациях и перестановках.

## Закон сложения в комбинаторике

Допустим, что есть две группы: в одной $n$ различных элементов, во второй $k$ различных элементов. Если из первой группы какой-либо элемент можно выбрать $n$ способами, а из второй - $k$ способами, то выбрать один элемент из первой или второй группы можно $n+k$ способами.

> Закон сложения используется тогда, когда нужно выбрать только `1` элемент.

**Пример**. В вазе лежат 5 яблок, 4 груши и 3 мандарина. Сколько существует способов взять один фрукт из вазы?

**Решение**. Один фрукт из вазы можно взять 12 способами: $5+4+3=12$

## Закон умножения в комбинаторике

Если элемент $A$ можно выбрать $n$ и затем второй элемент $B$ можно выбрать $k$ различными способами, то пару элементов $A$ и $B$ можно выбрать $n\cdot k$ способами.

**Пример**. У студента имеются рубашки 4 цветов и 3 вида галстука а также белые и чёрные брюки. Сколько всего вариантов подбора одежды есть у студента?  
**Решение**. У студента есть 24 варианта подбора одежды: $4\cdot3\cdot2=24$

## Перестановка

**Перестановка n объектов/элементов** — это способ их последовательного расположения с учётом порядка.Например, abc, bca и cab — это разные перестановки трёх букв.  
$$P_n = n!$$  
**Пример**. На странице интернет-магазина одежды размещены три футболки. Если поменять их расположение на странице, получится новая перестановка. Сколькими способами можно расположить футболки на странице?  
**Решение**. Три футболки можно расположить на странице 6 способами: $P_3=3!=1\cdot2\cdot3$

## Перестановка с повторяющимися элементами

**Перестановка с повторяющимися элементам** - когда нужно последовательно расположить элементы, некоторые из которых могут повторяться.

$$\overline{P_n}=\frac{n!}{n_1!\cdot n_2!\cdot ...\cdot n_k!}$$

**Пример**. Сколько различных пятибуквенных слов можно составить из слова МАННА?  
**Решение**. В слове МАННА буквы "а" и "н" повторяются 2 раза, а буква "м" 1 раз. Следовательно:  
$$\overline{P_5}=\frac{5!}{2!\cdot2!\cdot1!}=\frac{1\cdot2\cdot3\cdot4\cdot5}{4}=30$$

```python
num_balls = 7
num_reds = 3
num_blues = 2
num_greens = 2
num_variants = math.factorial(num_balls) / (math.factorial(num_reds) * math.factorial(num_blues) * math.factorial(num_greens))
print(num_variants)
```

## Размещение

**Размещение из n по k** — это упорядоченный набор из k различных элементов, взятых из некоторого множества с мощностью n, где $\text{k} \leq \text{n}$. То есть некая перестановка k выбранных элементов из n.  
$$A_n^k = \frac{n!}{(n-k)!}$$  
**Пример**. Сколькими способами можно заполнить спортивный пьедестал из трёх мест, если есть 10 претендентов?  
**Решение**.  
$$\frac{10!}{(10 - 3)!} = \frac{10!}{7!} = \frac{7! \cdot 8 \cdot 9 \cdot 10}{7!} = 720$$

## Размещение с повторяющимися элементами

**Размещение с повторяющимися элементами** - когда во множестве n один и тот же элемент можно взять несколько раз.  
$$\overline{A}_n^k=n^k$$  
**Пример**. Замок чемодана состоит из трёх дисков, каждый из которых имеет значения от 0 до 9. Сколько комбинаций паролей нужно перебрать чтобы подобрать правильный пароль к чемодану?

$$10^3=1000$$

## Сочетание

Когда порядок выбора или расположения не важен, говорят о **сочетании**.  

**Сочетание из n по k** — это неупорядоченный набор из k различных элементов, взятых из некоторого множества с мощностью n, где $\text{k} \leq \text{n}$. То есть набор, для которого порядок выбора не имеет значения.  
$$C_n^k  = \frac{n!}{k! \cdot (n - k)!}$$  
Порядок выбора или расстановки не важен, когда:  
* выбирают несколько элементов одновременно (например, когда из мешка с шариками вытаскивают несколько шариков разом)  
* выбирают пару или группу для взаимного или равноправного процесса (например пять человек для хора)

**Пример**. Из 9 актёров выбирают четырёх для массовки. Порядок выбранных людей не важен. Сколько есть способов выбрать актёров?  
**Решение**  
$$C_9^4=\frac{9!}{4! \cdot (9-4)!}=\frac{9!}{4! \cdot 5!}=\frac{5!\cdot6\cdot7\cdot8\cdot9}{1\cdot2\cdot3\cdot4\cdot5!}=\frac{3024}{24}=126$$

## Сочетание с повторяющимися элементами

**Сочетание с повторяющимися элементами** - когда порядок элементов не важен, но при этом один и тот же элемент можно выбрать несколько раз.

$$\overline{C}_9^4=\frac{(n+k-1)!}{k! \cdot (n-1)!}$$

**Пример**. На прилавке размещены лотки с пятью разными вкусами мороженного. Мы  можем взять три ложки из трёх разных лотков, либо все три из одного лотка, либо две из одного лотка а одну из другого. Сколько всего есть способов выбрать три ложки мороженного?

$$\overline{C}_5^3=\frac{(5+3-1)!}{3!\cdot(5-1)!}=\frac{7!}{3!\cdot4!}=\frac{4!\cdot5\cdot6\cdot7}{1\cdot2\cdot3\cdot4!}=35$$

❓Задача  
В цветочном магазине продаётся четыре цвета роз. Покупатель хочет собрать букет из 10 роз. Они могут быть как одного цвета так и разных цветов. Сколькими способами покупатель может собрать букет?

```python
num_of_colors = 4
num_of_picks = 10

nominator = math.factorial(num_of_colors + num_of_picks - 1)
denominator = math.factorial(num_of_picks) * math.factorial(num_of_colors - 1)

num_of_variants = nominator / denominator
num_of_variants
```

## Шпаргалка

|   | с повторами | без повторов |
|---|---|---|
| **порядок важен** | $n^k$ | $\frac{n!}{(n-k)!}$ |
| **порядок не важен** | $\frac{(n+k-1)!}{k!(n-1)!}$ | $\frac{n!}{k!(n-k)!}$ |

```python
p_threeH = num_of_desired_outcomes / num_of_possible_outcomes
p_threeH
```
