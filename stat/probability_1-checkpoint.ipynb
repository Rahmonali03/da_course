**Теория вероятности** - раздел математики, который изучает закономерности случайных явлений: случайные события, случайные величины, их свойства и операции над ними.  
Иными словами, теория вероятности помогает нам понять какова вероятность что то или иное событие произойдёт. Например:  
* какова вероятность того что при подбрасывании монеты десять раз подряд будет выпадать орёл?
* при броске двух игральных костей, какова вероятность что выпадут две шестёрки?
* какова вероятность того что пять вытащенных из колоды карт все будут червями?
* каковы ваши шансы выиграть в лотерею?

## Пространство событий

{H, T}

**Пространство элементарных событий** (S) - множество всех возможных исходов случайного эксперимента. Например:  
* при подбрасывании монеты возможны 2 исхода - орёл (H) либо решка (T). Следовательно пространство элементарных событий этого эксперимента `{H, T}`
* если же монету подбросить дважды то пространство элементарных событий будет уже состоять из четырёх возможных исходов `{HH, HT, TH, TT}`
* если подбросить монету трижды, то пространство элементарных событий будет следующим: `{HHH, THH, HTH, HHT, HTT, THT, TTH, TTT}`
* если же наш эксперимент заключается в подбрасывании монеты до тех пор пока не выпадет решка, то у нас будет бесконечное пространство элементарных событий. Т.е. мы можем получить решку после первого броска, либо после второго, либо после третьего и т.д. `{T, HT, HHT, HHHT, ...}`

>❗Сумма всех исходов пространства событий равна 1

**Случайное событие** - подмножество множества исходов случайного эксперимента; при многократном повторении случайного эксперимента частота наступления события служит оценкой его вероятности.

Одним из предыдущих примеров было подбрасывание монеты три раза. Пространство элементарных событий в таком эксперименте выглядело следующим образом: `S = {HHH, THH, HTH, HHT, HTT, THT, TTH, TTT}`.  
Случайными событиями для этого эксперимента могут быть:  
* **Событие A** - выпадение орла 0 раз  
* **Событие B** - выпадение орла лишь один раз   
* **Событие C** - выпадение орла хотя бы один раз  

Таким образом, каждое из указанных событий является подмножеством множества исходов:  
* **Событие A** --> `TTT`  
* **Событие B** --> `HTT, THT, TTH`  
* **Событие C** --> `HTT, THT, TTH, THH, HTH, HHT, HHH`  
{FBC58DD5-15E1-427E-9641-28630064E9FE}.png

Как видно на рисунке выше, `событие B` находится внутри `события C`, так как исходы относящиеся к `событию B`, также являются частью подмножества исходов `события C`. А `событие A` стоит отдельно так как оно никак не пересекается с другими событиями.

**Случайная величина** - переменная, значения которой представляют собой численные исходы некоторого случайного феномена или эксперимента. Другими словами, это численное выражение результата случайного события.

P(A)

Виды событий:
* **достоверные** - те, которые точно произойдут. Если бросить стакан на пол, то с вероятностью 100% он полетит вниз.
* **невозможные** - те, которые никогда не произойдут. Если бросить тот же стакан на пол, то он никогда не полетит вверх. Т.е. вероятность такого события 0%
* **независимые** - те, на которые не влияют другие события  
* **зависимые** - те, на которые влияют другие события  
* **несовместимые** - те, которые исключают друг-друга. Например, при подбрасывании монетки может выпасть либо орёл, либо решка - оба одновременно они выпасть не могут  
* **равновероятные** - два или более события, которые имеют одинаковые шансы произойти

>Таким образом вероятность любого события лежит в диапазоне от 0 до 1, где 0 означает что событие невозможно а 1 означает что событие точно произойдёт.$$0 \leq P(A) \leq 1$$

## Определение вероятности событий

Существует 2 основных метода определения вероятности какого-либо события:
* теоретический (также называемый классическим)
* эмпирический

**Классический** метод применим к тем случаям, когда каждый результат имеет равную вероятность, либо когда вероятность каждого результата заранее известна.
$$ P(A) = \frac{число\spaceблагоприятных\spaceисходов}{число\spaceвозможных\spaceисходов} $$

Например:  

*Подбрасывание монеты*  
У монеты 2 стороны - "орёл" и "решка". У "правильной" монеты вероятность получения орла или решки одинакова. Т.е. `P(орёл) = P(решка) = 1/2`. Таким образом `P(орёл) = 0.5`


*Игральная кость*  
У обычной игральной кости 6 сторон, и вероятность выпадания каждой из сторон является одинаковой. Т.е. `P(1) = P(2) = P(3) = P(4) = P(5) = P(6) = 1/6`


**Эмпирический** метод предполагает проведение многочисленных экспериментов для определения вероятности какого-либо события. Т.е. проводится `n`-ное количество экспериментов (например подбрасывание монеты) и результат каждого эксперимента фиксируется. Далее подсчитывается количество результатов соответствующих определённому событию (например выпадение орла), и этот результат (m) делится на количество экспериментов (n).  
$$ P_n(A) = \frac{m}{n} $$

Например:  

*Вероятность брака в производстве*  
При производстве какой-либо продукции на выходе получается некоторое количество брака. Бизнесу необходимо знать вероятность получения бракованной продукции чтобы в будущем заложить потери от брака в цену продукции. Эта вероятность рассчитывается делением количества брака на общее количество выпущенного продукта.  

*Нечестная монета*
Предположим что ваш друг 10 раз подбрасывает монету и каждый раз выпадает орёл. Возможно он использует специальную монету со смещённым центром тяжести и следовательно вероятность выпадания орла у такой монеты больше чем у обычной. Чтобы проверить это предположение монета подбрасывается достаточно большое количество раз и количество событий когда выпадал орёл делится на общее количество подбрасываний. Если полученный результат близок к 0,5 значит монета обычная, если нет значит это "нечестная" монета.
